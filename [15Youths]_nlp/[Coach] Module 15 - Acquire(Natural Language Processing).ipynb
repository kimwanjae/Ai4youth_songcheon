### 5.4 indexing

You can find the location of a particular word in the text when it first occurs<br>
텍스트가 처음 나타날 때 특정 단어의 위치를 찾을 수 있습니다

You can also use list comprehension to find out the positions of more matches in the list. Look at an [example](https://www.datacamp.com/community/tutorials/python-list-index) below:<br>
목록 이해를 사용하여 목록에서 더 많은 일치 항목의 위치를 찾을 수도 있습니다. 아래 예를 보세요.<br>
리스트에서는 매번 첫번째 값을 찾은다음 반환하고 종료하니까,,, list conprehension을 이용해 모든 3의 index 값을 찾아 표시하도록 

```python
list_numbers = [3, 1, 2, 3, 3, 4, 5, 6, 3, 7, 8, 9, 10]
[i for i, n in enumerate(list_numbers) if n == 3] # List comprehension
```
