### x 축과 y 축에 레이블 지정

## 서브플롯

종종 여러 그래프를 나란히 그려야할 때가 있습니다. 당신은 subplots()을 사용하여 그릴 수 있습니다.

여러 그래프를 나란히 그릴 수 있는 방법은 다음과 같습니다!

```python
x = dataframe.date
y = dataframe.mean_temperature

# y축을 공유하는 2개의 서브플롯 생성
fig, (ax1, ax2) = plt.subplots(1, 2, sharey=True)
fig.suptitle('Sharing Y axis')
ax1.plot(x, y)
ax2.scatter(x, y)
plt.show()
```

### Task 18:  y 축을 공유하는 3 개의 플롯을 아래에 그려봅시다!
